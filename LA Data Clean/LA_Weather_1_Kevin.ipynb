```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
la_weather_df = pd.read_csv("LA_Weather_15-17_Labels.csv")
la_crime_df = pd.read_csv("LA_Crime_Category_2015-2017.csv")
```

```python
la_weather_df.head()
```

```python
la_crime_df.head()
```

```python
la_crime_pivot = la_crime_df.groupby(["Date Occurred","Category"]).count()["Year"].reset_index().pivot("Date Occurred", "Category","Year")
```

```python
la_crime_pivot.head()
```

```python
la_crime_pivot.to_csv("LA_Crime_15-17_Pivot_1.csv")
```

```python
la_crime_pivot_date = la_crime_pivot.reset_index()
```

```python
la_crime_pivot_date["Date Occurred"] = la_crime_pivot_date["Date Occurred"].map(pd.to_datetime)
```

```python
la_crime_pivot_new = la_crime_pivot_date.sort_values("Date Occurred").rename(columns={"Date Occurred":"Date"}).set_index("Date")
```

```python
la_crime_pivot_new.to_csv("LA_Crime_15-17_Pivot_2.csv")
```

```python
la_weather_date = la_weather_df.copy()
la_weather_date["Date"] = la_weather_date["Date"].map(pd.to_datetime)
```

```python
la_weather_date
```

```python
la_weather_date.set_index("Date").to_csv("LA_Weather_15-17_Labels_Date.csv")
```

```python
chi_weather_df = pd.read_csv("Chicago_Weather_15-17_Labels.csv")
chi_weather_df.head()
```

```python
chi_weather_df["Date"] = chi_weather_df["Date"].map(pd.to_datetime)
```

```python
chi_weather_df
```

```python
chi_weather_df["Event"] = chi_weather_df["Event"].fillna("Clear")
```

```python
chi_weather_df
```

```python
chi_weather_df.set_index("Date").to_csv("Chicago_Weather_15-17_Labels_Date.csv")
```

```python
chi_crime_df = pd.read_csv("Chicago_Crime_2015-2017_UPDATE.csv")
chi_crime_df.head()
```

```python
chi_crime_df["Date"] = chi_crime_df["Date"].map(lambda x: x[:10])
```

```python
chi_crime_pivot = chi_crime_df.groupby(["Date","Category"]).count()["Year"].reset_index().pivot("Date", "Category","Year")
chi_crime_pivot
```

```python
chi_crime_pivot_date = chi_crime_pivot.reset_index()
chi_crime_pivot_date["Date"] = chi_crime_pivot_date["Date"].map(pd.to_datetime)
chi_crime_pivot_date = chi_crime_pivot_date.sort_values("Date")
chi_crime_pivot_date
```

```python
chi_crime_pivot_date.set_index("Date").to_csv("Chicago_Crime_15-17_Pivot_1.csv")
```

```python
la_weather_1 = pd.read_csv("LA_Weather_15-17_Labels_Date.csv")
la_weather_1.head()
```

```python
la_weather_c = la_weather_1[["Date", "Temp (F) avg", "Event"]]
la_weather_c
```

```python
la_weather_c.to_csv("LA_Weather_15-17_Clean.csv")
```

```python
la_crime_pivot_1 = pd.read_csv("LA_Crime_15-17_Pivot_2.csv")
la_crime_pivot_1["Total"] = la_crime_pivot_1.iloc[:,1:].apply(sum, axis=1)
la_crime_pivot_1
```

```python
la_crime_pivot_1.set_index("Date").to_csv("LA_Crime_15-17_Pivot_Total.csv")
```

```python
la_merged_df = la_weather_c.merge(la_crime_pivot_1, on="Date", how="outer")
la_merged_df.to_csv("LA_15-17_Merged.csv")
```
